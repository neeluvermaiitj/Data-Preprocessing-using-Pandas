# Data Preprocessing

<font color="red"><h2>**Measurement scales**</h2></font>

<font><h3>*Non-metric Scale*</h3></font>
<font><h6>1.Nominal</h6></font>
<font><h6>2.Ordinal</h6></font>

<font><h3>*Metric Scale*</h3></font>
<font><h6>1.Interval</h6></font>
<font><h6>2.Ratio</h6></font>

<font><h2>**Nominal**</h2></font>
<font><h6> Charcteristics : A scale used for labeling variables into distinct classifications , in no order.</h6></font>
<font><h6>Examples: Name of Places, Brand Types</h6></font>

<font><h2>**Ordinal**</h2></font>
<font><h6>Charcteristics : It depicts the order of variables and not the difference between each of the variables.</h3></font>
<font><h6>Examples: Grades, Quality</h4></font>

<font><h2>**Interval**</h2></font>
<font><h6>Charcteristics : It indicates difference between the objects</h3></font>
<font><h6>Examples:Temperature</h4></font>

<font><h2>**Ratio**</h2></font>
<font><h6>Charcteristics : A measurement scale that produces the order of variables,also makes the difference between variables known along with information on the value of true zero.</h3></font>
<font><h6>Examples: Height , Weight</h4></font>

## Missing Values

### Read the csv file

```python
import pandas as pd

chemical = pd.read_csv("ml/data/ChemicalProcess.csv")
```

```python
chemical.head()
```

```python
#Check for missing value
print(chemical.isnull())
```

```python
chemical.describe()
```

```python
#Determine columns with missings
chemical.loc[:, chemical.isnull().any()].columns
```

```python
#Viewing the column names
chemical.columns
```

```python
#Information about the data
chemical.info()
```

### Dropping NA values

```python
chemicalDropNA = chemical.dropna(axis=1)
```

```python
chemicalDropNA.shape
```

```python
chemicalDropNA.info()
```

```python
#droppping selected columns
chemical_df=chemical.drop(["Unnamed: 44","Unnamed: 45"],axis=1)
```

```python
chemical_df.info()
```

### Replacing the missing values

```python
chemical_fill_na= chemical_df.fillna(0)
```

```python
chemical_fill_na
```

```python
#OR
from sklearn.impute import SimpleImputer
imp = SimpleImputer(strategy='mean')
```

```python
chemicalImputed = imp.fit_transform(chemical_df)
```

```python
chemicalImputed
```

```python
df_chemicalImputed = pd.DataFrame(chemicalImputed,columns= chemical_df.columns)
```

```python
df_chemicalImputed
```
